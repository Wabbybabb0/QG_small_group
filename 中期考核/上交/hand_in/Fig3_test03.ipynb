```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.linspace(0,20,2000)
data
```

```python
def opinion_update(X_t,size):
    """
    输入：
    X_t        ndarray(n,) 
    size       scaler
    
    输出：
    X_         ndarray(n,)
    """
    m = X_t.shape[0]
    X_ = np.zeros(size)
    
    for i in range(m):
        neighbor = 0
        times = 0
        for j in range(m):
             if(np.abs(X_t[i] - X_t[j]) < 1):
                neighbor += X_t[j]
                times += 1
        X_[i] = neighbor/times
    return X_

```

```python
y_0 = np.zeros((2000,16))
X_func = np.zeros((2000,16))
X_func[:,0] = data
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_0[:,0],X_func[:,0],s=0.1)
```

```python
now = time.time()
for i in range(1,16):
    y_0[:,i] = y_0[:,i] + i
    for j in range(2000):
        print(i,j,'耗时:',time.time()-now,'s')
        now = time.time()
        X_func[:,i] = opinion_update(X_func[:,i-1], 2000)
        plt.scatter(y_0[j][i],X_func[j][i],s=1)
        plt.plot(y_0[j],X_func[j])
```
