```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#创建数据集

data_1 = np.linspace(0,1,300)
data_2 = np.linspace(3,4,300)
data_3 = np.linspace(1,3,150)
data = np.concatenate((data_1,data_3,data_2),axis=0)
data.shape
```

```python
def opinion_update(X_t,size):
    """
    输入：
    X_t        ndarray(n,) 
    size       scaler
    
    输出：
    X_         ndarray(n,)
    """
    m = X_t.shape[0]
    X_ = np.zeros(size)
    
    for i in range(m):
        neighbor = 0
        times = 0
        for j in range(m):
             if(np.abs(X_t[i] - X_t[j]) < 1):
                neighbor += X_t[j]
                times += 1
        X_[i] = neighbor/times
    return X_
```

```python
y_0 = np.zeros((750,16))
X_fuc = np.zeros((750,16))
X_fuc[:,0]=data
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_0[:,0],X_fuc[:,0],s=1)
```

```python
for i in range(1,16):
    y_0[:,i] = y_0[:,i] + i
    for j in range(750):
        print(i,j)
        X_fuc[:,i] = opinion_update(X_fuc[:,i-1], 750)
        plt.scatter(y_0[j][i],X_fuc[j][i],s=1)
        plt.plot(y_0[j],X_fuc[j])
```
